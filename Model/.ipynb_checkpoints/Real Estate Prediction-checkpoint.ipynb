```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20, 10)
pd.options.mode.chained_assignment = None
```

```python
df1 = pd.read_csv('Bengaluru_House_Data.csv')
df1.head()
```

# Data Preprocessing

```python
df2 = df1.drop(['area_type', 'balcony', 'availability', 'society'], axis=1)
df3 = df2.dropna()

df3['bhk'] = df3['size'].apply(lambda x : int(x.split(' ')[0]))
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

def convert_sqft(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
        
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft)
# df3[~df3['total_sqft'].apply(is_float)]
```

```python
df5 = df4.copy()
df5['price_per_sqft'] = (df5.price*100000) / df5.total_sqft
df5.head()
```

```python
df5.location = df5.location.apply(lambda x : x.strip())

location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending= False)
location_stats

location_other = location_stats[location_stats <= 10]
df5.location = df5.location.apply(lambda x: 'other' if x in location_other else x)

df5.head()
```

# Outlier Removal

```python
df6 = df5[~(df5.total_sqft / df5.bhk <300)]
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        std = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft<(m+std)) & (subdf.price_per_sqft>=(m-std))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df7 = remove_pps_outliers(df6)

df7.shape
```

```python
def plot_scatter_plot(df, location):
    bhk2 = df[(df.location==location) & (df.bhk==2)]
    bhk3 = df[(df.location==location) & (df.bhk==3)]
    
    plt.scatter(bhk2.total_sqft, bhk2.price, color='blue', label='2 bhk', s=50)
    plt.scatter(bhk3.total_sqft, bhk3.price, color='red', marker='+', label='3 bhk', s=50)
    
    plt.xlabel('Area')
    plt.ylabel('Price')
    plt.title(location)
    plt.legend()
    
# plot_scatter_plot(df7, 'Hebbal')
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std' : np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < stats['mean']].index.values)
    return df.drop(exclude_indices, axis='index')
```

```python
df8 = remove_bhk_outliers(df7)

# plot_scatter_plot(df8, 'Hebbal')
```

```python
df9 = df8[df8.bath < df8.bhk+2]
df9.shape
```

```python
df10 = df9.drop(['size', 'price_per_sqft'], axis=1)
df10.head()
```

```python
dummy = pd.get_dummies(df10.location)

df11 = pd.concat([df10, dummy.drop('other', axis='columns')], axis='columns')

df11.head()
```

```python
df12 = df11.drop(['location'], axis='columns')
df12.head()
```

# Model Selection

```python
X = df12.drop(['price'], axis='columns')
y = df12.price
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import train_test_split

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
```

```python
cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
cross_val_score(Lasso(), X, y, cv=cv)
```

```python
cross_val_score(DecisionTreeRegressor(), X, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

model_params = {
    'LinearRegression': {
        'model': LinearRegression(),
        'params': {
            
        }
    },
    'Lasso': {
        'model': Lasso(),
        'params': {
            'alpha': [0.5, 1, 1.5, 2]
        }
    },
    'DecisionTreeRegressor': {
        'model': DecisionTreeRegressor(),
        'params': {
            'criterion': ['mse', 'mae'],
            'splitter': ['best', 'random']
        }
    }
}

scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=cv, return_train_score=False)
    clf.fit(X, y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

rank = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
```

```python
rank.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test,  y_test)
```

```python
np.where(X.columns == 'Rajaji Nagar')
```

```python
def predict_price(location, sqft, bath, bhk):
    loc_index = np.where(X.columns == location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1
    return lr.predict([x])[0]
    
```

```python
predict_price('Indira Nagar', 1000, 2, 2)
```
